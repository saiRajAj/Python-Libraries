Working with Time Series

DateTime Index

```python
import pandas as pd
import numpy as np

np.random.seed(0)

dates = pd.date_range("2023-01-01", "2023-12-31")
customers = ["Alice", "Bob", "David", "Frank", "Helen", "Ivy"]

df = pd.DataFrame({
    "Date": np.random.choice(dates, 100),
    "Customer Name": np.random.choice(customers, 100),
    "Lemonades": np.random.randint(1, 11, 100)
})

print(df)
```

```python
df
```

```python
df["Date"]=pd.to_datetime(df["Date"])
```

```python
df
```

```python
df.set_index("Date", inplace=True)
```

```python
df
```

Convert string to dates

```python
pd.to_datetime("2023-01-01")
```

Resampling

```python
df.resample('W').sum()
```

```python
df.resample('A').sum()
```

Shifting/laging

```python
df['Previous Lemonades'] = df['Lemonades'].shift(1)
```

```python
df
```

Rolling Calculations

```python
df['Rolling Mean'] = df['Lemonades'].rolling(window=3).mean()
```

```python
df
```
